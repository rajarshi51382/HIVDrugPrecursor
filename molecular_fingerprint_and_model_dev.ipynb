```python
df_ccr5_X = pd.read_csv('descriptors_output.csv')
df_ccr5_X = df_ccr5_X.drop(columns=['Name'])
df_ccr5_X

df_kat5_X = pd.read_csv('descriptors_output.csv')
df_kat5_X = df_kat5_X.drop(columns=['Name'])
df_kat5_X

df_protease_X = pd.read_csv('descriptors_output.csv')
df_protease_X = df_protease_X.drop(columns=['Name'])
df_protease_X
```

```python
df_ccr5_Y=df_ccr5['pIC50']
df_ccr5_Y

df_kat5_Y=df_kat5['pIC50']
df_kat5_Y

df_protease_Y=df_protease['pIC50']
df_protease_Y
```

```python
dataset_ccr5=pd.concat([df_ccr5_X,df_ccr5_Y], axis=1)
dataset_ccr5

dataset_kat5=pd.concat([df_kat5_X,df_kat5_Y], axis=1)
dataset_kat5

dataset_protease=pd.concat([df_protease_X,df_protease_Y], axis=1)
dataset_protease
```

```python
dataset_ccr5.to_csv('ccr5_pubchem_fp.csv', index=False)
dataset_protease.to_csv('protease_pubchem_fp.csv', index=False)
dataset_kat5.to_csv('kat5_pubchem_fp.csv', index=False)
```
